```python
DATASET_FOLDER = "/home/jovyan/work/data/waymo-e2e/waymo_open_dataset_end_to_end_camera_v_1_0_0/"

TRAIN_FILES = glob.glob(os.path.join(DATASET_FOLDER, "training*.tfrecord*"))
VALIDATION_FILES = glob.glob(os.path.join(DATASET_FOLDER, "val*.tfrecord*"))
TEST_FILES = glob.glob(os.path.join(DATASET_FOLDER, "test*.tfrecord*"))

print("Train:", len(TRAIN_FILES), "Validation:", len(VALIDATION_FILES), "Test:", len(TEST_FILES))

```

```python
filenames = tf.io.matching_files(VALIDATION_FILES)
dataset = tf.data.TFRecordDataset(filenames, compression_type='')
dataset_iter = dataset.as_numpy_iterator()
```

```python
dataset_iter = iter(dataset)
try:
    bytes_example = next(dataset_iter)
    print(bytes_example)
except StopIteration:
    print("Dataset is empty!")
```

### Iteration Approach Breakdown

```python
raw_record = next(iter(tf.data.TFRecordDataset(VALIDATION_FILES)))
ex = tf.train.Example()
ex.ParseFromString(raw_record.numpy())

print("Available feature keys:")
for k in ex.features.feature.keys():
    print(" -", k)
```

```python
raw_record = next(iter(tf.data.TFRecordDataset(VALIDATION_FILES)))
ex = tf.train.Example()
ex.ParseFromString(raw_record.numpy())

feat = ex.features.feature["21c8f36a1a5a4075a7439ba47b302a89-040"]

print("bytes_list length:", len(feat.bytes_list.value))
print("int64_list length:", len(feat.int64_list.value))
print("float_list length:", len(feat.float_list.value))

# If bytes exist, print size of first element
if len(feat.bytes_list.value) > 0:
    print("First bytes length:", len(feat.bytes_list.value[0]))
```
